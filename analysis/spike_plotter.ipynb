```python
import brian2
import csv
import matplotlib
import matplotlib.pyplot as plt
```

```python
def csv_to_raster(file_name,plot_title):
   with open(file_name+'.csv', 'r') as csv_file:
    data = list(csv.reader(csv_file, delimiter=','))
    spike_idx = [int(idx) for idx in data[0]]
    spike_times = [float(time[:-2]) for time in data[1]] # remove last 2 charactes from times as these will be 'ms' to signify milliseconds
    fig = plt.figure(dpi=150)
    matplotlib.rcParams['figure.figsize'] = [6.5, 4]
    plt.plot(spike_times, spike_idx, 'ok', markersize=0.5)
    plt.xlabel('Time (ms)')
    plt.ylabel('Neuron index')
    plt.title(plot_title)
    # plt.savefig(file_name+'.png',dpi=500)
```

```python
csv_to_raster('../simulation_data/layer_1_excitatory_spikes','Layer 1 excitatory neurons')
```

```python
csv_to_raster('simulation_data/layer_1_inhibitory_spikes','Layer 1 inhibitory neurons')
```
